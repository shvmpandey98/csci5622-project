
## Colorado Stream Flow Data

```python
# Importing the necessary libraries
import dataretrieval.nwis as nwis
import pandas as pd
import matplotlib.pyplot as plt
```

### Fetching the data from the USGS NWIS via API call using the dataretrieval library

```python
# UNCOMPAHGRE RIVER NEAR OURAY, CO
# USGS URL: https://dashboard.waterdata.usgs.gov/api/gwis/2.1/service/site?agencyCode=USGS&siteNumber=09146200&open=211902

site = '09146020'  # Uncompahgre River near Ouray, CO

display(nwis.get_record(sites=site, service='site')) # Site information

# get instantaneous values (iv)
df_uncompahgre_river = nwis.get_record(sites=site, service='dv', start='2009-01-01', end='2024-02-08').reset_index()

display(df_uncompahgre_river.head()) # Raw data

df_uncompahgre_river = df_uncompahgre_river[['datetime', '00060_Mean']]
df_uncompahgre_river.columns = ['datetime', 'daily_discharge_cfs']

df_uncompahgre_river['datetime'] = df_uncompahgre_river['datetime'].dt.date # Extracting date from datetime

display(df_uncompahgre_river.head())
```

#### Uncompahgre River near Ouray, CO Streamflow Data Analysis

```python
# Summary of the data for Uncompahgre River near Ouray, CO
display(df_uncompahgre_river.describe())

# Here -999999 is the missing value. We will remove those rows.
df_uncompahgre_river = df_uncompahgre_river[df_uncompahgre_river['daily_discharge_cfs'] != -999999]

# Summary of the data for Uncompahgre River near Ouray, CO after removing missing values
display(df_uncompahgre_river.describe())
```

```python
# Plotting time series of daily discharge for Uncompahgre River near Ouray, CO

plt.figure(figsize=(15, 6))
plt.plot(df_uncompahgre_river['datetime'], df_uncompahgre_river['daily_discharge_cfs'], label='Uncompahgre River near Ouray, CO')
plt.xlabel('Date')
plt.ylabel('Daily Discharge (cfs)')
plt.title('Daily Discharge of Uncompahgre River near Ouray, CO')
plt.legend()
plt.show()
```

```python
# Plotting distribution of daily discharge for Uncompahgre River near Ouray, CO

plt.figure(figsize=(15, 6))
plt.hist(df_uncompahgre_river['daily_discharge_cfs'], bins=50, alpha=0.7, color='blue')
plt.xlabel('Daily Discharge (cfs)')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Discharge of Uncompahgre River near Ouray, CO')
plt.show()
```
