```python
# Importing the libraries

import pandas as pd
import numpy as np
```

```python
# Reading the final dataset created in previous ipynb files

df_final = pd.read_parquet('data/final_data/pueblo_pca.parquet')

display(df_final.head())

df_final.shape
```

```python
# Reading target data

df_target = pd.read_parquet('data/final_data/target.parquet')

df_target.head()
```

```python
# Dropping D0, D1, D1 and None columns, as we are only considering severe drought and above conditions
df_target = df_target.drop(['D0', 'D1', 'D2', 'D3', 'None'], axis=1)

# Dropping the rows where D4 is 0
df_target = df_target[(df_target['D4'] != 0)]

# Creating a new column 'target' with all value as 1
df_target['target'] = 1 # 1 for drought

# Droping D4 columns
df_target = df_target.drop(['D4'], axis=1)

df_target.head()
```

```python
# Merging the final dataset with target dataset

df_final = df_final.merge(df_target, on='date', how='left')
```

```python
# Replacing NaN values in target column with 0
df_final['target'] = df_final['target'].fillna(0) # 0 for no drought

# Converting the target column to integer
df_final['target'] = df_final['target'].astype(int)

# Selecting the features and target
final_columns = ['date', 'precip', 'max_temp', 'min_temp', 'gwl', 'daily_discharge_cfs', 'sm_pca', 'st_pca', 'target']

df_final = df_final[final_columns]
```

```python
# Saving the final dataset

df_final.to_parquet('data/final_data/final_dataset.parquet', index=False)
```
