## Colorado Groundwater Data

```python
# Importing required libraries
import dataretrieval.nwis as nwis
import pandas as pd
import matplotlib.pyplot as plt
```

### Fetching the data from the USGS NWIS via API call using the dataretrieval library

```python
# Muddy Creek Well, CO
# USGD URL: https://dashboard.waterdata.usgs.gov/api/gwis/2.1/service/site?agencyCode=USGS&siteNumber=393902103554001&open=212456

site = '393902103554001' # Site number for Muddy Creek Well, CO
display(nwis.get_record(sites=site, service='site'))


# get instantaneous values (iv)
df_gwl_muddy_creek = nwis.get_record(sites=site, service='dv', start='2009-01-01', end='2022-11-30').reset_index()

display(df_gwl_muddy_creek.head()) # Raw data

df_gwl_muddy_creek = df_gwl_muddy_creek[['datetime', '72019_Mean']]
df_gwl_muddy_creek.columns = ['datetime', 'gwl']
df_gwl_muddy_creek['datetime'] = df_gwl_muddy_creek['datetime'].dt.date # Extracting date from datetime


display(df_gwl_muddy_creek.head())
```

```python
# Drought Well, Near Pueblo
# USGD URL: https://dashboard.waterdata.usgs.gov/api/gwis/2.1/service/site?agencyCode=USGS&siteNumber=382323104200701&open=212313

site = '382323104200701' # Site number for Drought Well, Near Pueblo
display(nwis.get_record(sites=site, service='site'))

# get instantaneous values (iv)
df_gwl_pueblo = nwis.get_record(sites=site, service='dv', start='2009-01-01', end='2022-11-30').reset_index()

display(df_gwl_pueblo.head()) # Raw data

df_gwl_pueblo = df_gwl_pueblo[['datetime', '72019_Mean']]
df_gwl_pueblo.columns = ['datetime', 'gwl']
df_gwl_pueblo['datetime'] = df_gwl_pueblo['datetime'].dt.date # Extracting date from datetime


display(df_gwl_pueblo.head())
```

```python
# Kit Carson County, CO
# USGD URL: https://dashboard.waterdata.usgs.gov/api/gwis/2.1/service/site?agencyCode=USGS&siteNumber=390740103055801&open=212441

site = '390740103055801' # Site number for Kit Carson County, CO
display(nwis.get_record(sites=site, service='site'))


# get instantaneous values (iv)
df_gwl_kit_carson = nwis.get_record(sites=site, service='dv', start='2009-01-01', end='2022-11-30').reset_index()

display(df_gwl_kit_carson.head()) # Raw data

df_gwl_kit_carson = df_gwl_kit_carson[['datetime', '72019_Mean']]
df_gwl_kit_carson.columns = ['datetime', 'gwl']
df_gwl_kit_carson['datetime'] = df_gwl_kit_carson['datetime'].dt.date # Extracting date from datetime

display(df_gwl_kit_carson.head())
```

#### Muddy Creek Well, CO Groundwater Level Data Analysis

```python
# Summary of the data for Muddy Creek Well, CO
display(df_gwl_muddy_creek.describe())
```

```python
# Plotting time series for Muddy Creek Well, CO

plt.figure(figsize=(15, 6))
plt.plot(df_gwl_muddy_creek['datetime'], df_gwl_muddy_creek['gwl'], label='Muddy Creek Well, CO')
plt.xlabel('Date')
plt.ylabel('Groundwater Level (ft)')
plt.title('Time Series of Groundwater Level for Muddy Creek Well, CO')
plt.legend()
plt.show()
```

```python
# Plotting distribution of groundwater level for Muddy Creek Well, CO

plt.figure(figsize=(15, 6))
plt.hist(df_gwl_muddy_creek['gwl'], bins=30, alpha=0.5)
plt.xlabel('Groundwater Level (ft)')
plt.ylabel('Frequency')
plt.title('Distribution of Groundwater Level for Muddy Creek Well, CO')
plt.show()
```
